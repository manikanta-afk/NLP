```python
medical_corpus = """
Diabetes mellitus is a chronic metabolic disorder characterized by high blood glucose levels.
Hypertension increases the risk of cardiovascular diseases such as stroke and heart attack.
Antibiotics are prescribed to treat bacterial infections, but misuse leads to antimicrobial resistance.
Early diagnosis of cancer improves patient survival rates and treatment outcomes.
Vaccination is an effective method to prevent infectious diseases and reduce mortality.
"""
import nltk
nltk.download('stopwards')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
words_in_quote = word_tokenize(medical_corpus)
stop_words = set(stopwords.words("english"))
filtered_list = []
for word in words_in_quote:
    if word.casefold() not in stop_words:
        filtered_list.append(word)
filtered_list
```

```python
import nltk
nltk.download('punkt_tab')
medical_corpus = """
Diabetes mellitus is a chronic metabolic disorder characterized by high blood glucose levels.
Hypertension increases the risk of cardiovascular diseases such as stroke and heart attack.
Antibiotics are prescribed to treat bacterial infections, but misuse leads to antimicrobial resistance.
Early diagnosis of cancer improves patient survival rates and treatment outcomes.
Vaccination is an effective method to prevent infectious diseases and reduce mortality.
"""
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize
snowball = SnowballStemmer(language='english')
words = word_tokenize(medical_corpus)
for word in words:
    print(word,"---->",snowball.stem(word))
```

```python
from nltk import LancasterStemmer
Lanc = LancasterStemmer()
words = word_tokenize(medical_corpus)
for word in words:
    print(word,"---->",Lanc.stem(word))
```

```python
from nltk.stem import RegexpStemmer
regexp = RegexpStemmer('ing$|s$|e$|able$', min=4)
words = word_tokenize(medical_corpus)
for word in words:
    print(word,"---->",regexp.stem(word))
```

```python
medical_corpus = """
Diabetes mellitus is a chronic metabolic disorder characterized by high blood glucose levels.
Hypertension increases the risk of cardiovascular diseases such as stroke and heart attack.
Antibiotics are prescribed to treat bacterial infections, but misuse leads to antimicrobial resistance.
Early diagnosis of cancer improves patient survival rates and treatment outcomes.
Vaccination is an effective method to prevent infectious diseases and reduce mortality.
"""
import nltk
nltk.download('omw-1.4')
nltk.download('punkt_tab')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words = word_tokenize(medical_corpus)
for word in words:
    print(word,"---->",lemmatizer.lemmatize(word))
```
